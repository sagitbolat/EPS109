<font color="red" size="5" > (6 pts for correct definition for each function ) </font>

## (1) The Weierstrass polynomial of n'th order T$_n$(x) is defined in terms of lower-order polynimials:

## $T_0(x) = 1$

## $T_1(x) = x$

## $T_{n+1}(x) = 2xT_{n}(x) - 4T_{n-1}(x)$

## Into one diagram, plot the functions $T_3(x)$, $T_4(x)$ , and $T_5(x)$ over the interval 0 to 1.

```python
#There are no "Weierstrass" polynomials by the way.

def T0(x):
    return 1
def T1(x):
    return x
def T2(x):
    return 2*x*T1(x) - 4*T0(x)

def T3(x):
    return 2*x*T2(x) - 4*T1(x)

def T4(x):
    return 2*x*T3(x) - 4*T2(x)

def T5(x):
    return 2*x*T4(x) - 4*T3(x)

def T(x,n):
    if n == 0:
        return 1
    elif n == 1:
        return x
    else:
        return 2*x*T(x,n-1) - 4*T(x,n-2)

x = np.arange(0,1,0.01)
y3 = [T3(x_) for x_ in x]
y4 = [T4(x_) for x_ in x]
y5 = [T5(x_) for x_ in x]
# Or code recursively
y3 = [T(x_,3) for x_ in x]
y4 = [T(x_,4) for x_ in x]
y5 = [T(x_,5) for x_ in x]


plt.plot(x,y3,label='T3(x)')
plt.plot(x,y4,label='T4(x)')
plt.plot(x,y5,label='T5(x)')
```

# Question 5: Random Walk With Diagonal Moves <font color="red" size="5" > (30+15 pts) </font>

### (1) Using random numbers, generate a random walk in 2D with 5000 steps. Let your particle start walking from (x=0,y=0). You are only allowed to make diagonal moves. Each diagonal step should be taken with the same probability. Make an X-Y plot of the resulting series of steps. This also called a trajectory. (No 2D array is needed. Instead we recommend all X and Y coordinates to be stored in Python lists or arrays before being plotted.) <font color="red" size="5" > (15 pts) </font>

<font color="red" size="3" > (3 pts for generation of random number) </font>

<font color="red" size="3" > (4 $\times$ 1 pts for each correct move) </font>

<font color="red" size="3" > (8 pts for plotting the figure) </font>

```python
r0 = [0,0]
x0 = 0
y0 = 0
x = x0
y = y0
xs = []
ys = []
nn = 5000
for i in range(nn):
    r = np.random.random()
    if r<=0.25:
        x += 1
        y += 1
    elif r<=0.5:
        x += 1
        y -= 1
    elif r<=0.75:
        x -= 1
        y += 1
    else:
        x -= 1
        y -= 1
        
    xs.append(x)
    ys.append(y)
    
plt.plot(xs,ys)
plt.show()        
```

### (2) Now calculate the fraction of points on your trajectory that have been stepped on more than once. <font color="red" size="5" >(15 pts)</font>

<font color="red" size="3" > (5 pts for a loop tracing back each step) </font>

<font color="red" size="3" > (5 pts for finding overstepped positions) </font>

<font color="red" size="3" > (5 pts for correct answer) </font>

```python
boolean = np.zeros(nn)
for i in range(nn):
    xi = xs[i]
    yi = ys[i]
    for j in range(i):
        xj = xs[j]
        yj = ys[j]
        if xi == xj and yi == yj:
            boolean[i] = 1
            boolean[j] = 1 # needed of j>i is excluded in 'j' loop

print("number of stepped points = {}; fraction = {}".format(sum(boolean),sum(boolean)/nn))
     
```
