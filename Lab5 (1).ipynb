```python
def Mandlebrot(c, n_max, z_lim):
    z_curr = 0
    diverges = False
    for n in range(n_max):
        z_curr = ((z_curr * z_curr) + c)
        if abs(z_curr) >= z_lim:
            diverges = True
            break
    return diverges

print("C = -4 diverges?:        ", Mandlebrot(-4, 30, 50))
print("C = -2.001 diverges?:    ", Mandlebrot(-2.001, 30, 50))
print("C = -2 diverges?:        ", Mandlebrot(-2, 30, 50))
print("C = -1.999 diverges?:    ", Mandlebrot(-1.999, 30, 50))
print("C = -1 diverges?:        ", Mandlebrot(-1, 30, 50))
print("C = 0 diverges?:         ", Mandlebrot(0, 30, 50))
print("C = +0.5 diverges?:      ", Mandlebrot(0.5, 30, 50))
print("C = +1 diverges?:        ", Mandlebrot(1, 30, 50))
print("C = +2 diverges?:        ", Mandlebrot(2, 30, 50))
```

```python
import matplotlib.pyplot as plt
import numpy as np

def fc(c):
    return Mandlebrot(c, 30, 50)

c_data = np.arange(-4, 2, 0.001)

fc_data = np.array(list(map(fc, c_data)))

plt.plot(c_data, fc_data)
plt.title("C vs f(C)")
plt.xlabel("C")
plt.ylabel("f(C)")
plt.show()
```

```python
import math

def complex_abs(z):
    r = z.real
    i = z.imag
    re = r * (math.e)
    ie = i * (math.e)
    return np.sqrt(re*re + ie*ie)

def complex_mandlebrot(c, n_max, z_lim):
    z_curr = complex(0, 0)
    diverges = False
    for n in range(n_max):
        z_curr = ((z_curr * z_curr) + c)
        if complex_abs(z_curr) >= z_lim:
            diverges = True
            break
    return diverges



c_complex =np.array([
            complex(-0.4, 0.2),
            complex(0.4, 0.6),
            complex(-1.7, 0),
            complex(-1.4, 0.001),
            complex(-1.778, 0),
            complex(-1.778, 0.001),
            complex(-1.25, 0),
            complex(-1.25, 0.04),
            complex(-1.25, -0.9),
            complex(-1.25, -0.85)])


for c in c_complex:
    print("C =", c, "diverges?:", complex_mandlebrot(c, 30, 50))
```

```python
N = 201
xs = np.linspace(start=-2.0, stop=0.5, num=N)
ys = np.linspace(start=-1.25, stop=1.25, num=N)


A = np.zeros((N, N))

for i in range(N):
    for j in range(N):
        c = complex(xs[i], ys[j])
        A[j, i] = complex_mandlebrot(c, 50, 50)

plt.imshow(A)
```

```python
def complex_mandlebrot_ret_n(c, n_max, z_lim):
    z_curr = complex(0, 0)
    for n in range(n_max):
        z_curr = ((z_curr * z_curr) + c)
        if complex_abs(z_curr) >= z_lim:
            return n
    return n_max

N = 201
xs = np.linspace(start=-2.0, stop=0.5, num=N)
ys = np.linspace(start=-1.25, stop=1.25, num=N)


A = np.zeros((N, N))

for i in range(N):
    for j in range(N):
        c = complex(xs[i], ys[j])
        A[j, i] = complex_mandlebrot_ret_n(c, 50, 50)
        
plt.imshow(A)
```
