### 1.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def complex_abs(z):
    r = z.real
    i = z.imag
    re = r * (math.e)
    ie = i * (math.e)
    return np.sqrt(re*re + ie*ie)

def mandlebrot(z_init, c, n_max, z_lim):
    z_curr = z_init
    for n in range(n_max):
        z_curr = ((z_curr * z_curr) + c)
        if complex_abs(z_curr) >= z_lim:
            return n
    return n_max

N = 500
interval = 1.5
y_offset = 0
x_offset = 0
xs = np.linspace(start=-interval + x_offset, stop=interval + x_offset, num=N)
ys = np.linspace(start=-interval + y_offset, stop=interval + y_offset, num=N)


A = np.zeros((N, N))

for i in range(N):
    for j in range(N):
        x = xs[i]
        y = ys[j]
        z = complex(x, y)
        c = complex(-0.4, 0.6)
        A[j, i] = mandlebrot(z, c, 150, 50)
        
plt.imshow(A, origin ='lower')
plt.savefig("mandlebrot_1.png", dpi = N)
```
