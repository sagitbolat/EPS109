```python
n = 11
data = np.zeros((n,n))
# (a) what is the effect of the following three lines
data[1,1] = 1
data[2,3] = 1
data[3,5] = 1
# (b) again what do the following two lines do?
data[5,1:n-1] = 1
data[n-2,:] = 1
print(data)
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
# Here we practise 'for' loops
n = 11
data = np.zeros((n,n))
# (a) What is the effect of this 'for' loop?
for i in range(0,n):
    data[i,3] = 1
# (b) And this 'for' loop?
col = (n-1) // 2
for i in range(0,n,2):
    data[i,col] = 1
print(data)
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n,n))
for i in range(n):
    data[i, i] = 1
    data[10 - i, i] = 1
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n,n))
for i in range(1, n-1):
    data[1 , i] = 1
    data[n-2, i] = 1
    data[i, 1] = 1
    data[i, n-2] = 1
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n,n))

for j in range(0, 5, 2):
    k = n - j
    for i in range(j, k):
        data[0 , i] = 1
        data[k-1, i] = 1
        data[i, 0] = 1
        data[i, k-1] = 1
for i in range(6):
    data[2 + i, 2] = 1
    data[2, 2 + i] = 1
for i in range(2):
    data[4 + i, 4] = 1
    data[4, 4 + i] = 1
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n,n))
for i in range(1, n-1):
    for j in range(1, n-1):
        data[i, j] = 1
for i in range(n):
    data[i, 5] = 0
    data[5, i] = 0
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n,n))
data[5, 1] = 1
for i in range(3):
    data[4 + i, 3] = 1
for i in range(5):
    data[3 + i, 5] = 1
for i in range(7):
    data[2 + i, 7] = 1
for i in range(9):
    data[1 + i, 9] = 1
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 501
r = n/2.0
data = np.zeros((n,n))
for x in range(n):
    for y in range(n):
        x_diff = abs(251 - x)
        y_diff = abs(251 - y)
        x_2 = x_diff ** 2
        y_2 = y_diff ** 2
        distance_from_center = (x_2 + y_2) ** 0.5
        if (distance_from_center < r):
            data[x, y] = 1
plt.imshow(data, interpolation='nearest')
plt.show()
```
