```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 11
data = np.zeros((n,n))
for x in range(n):
    for y in range(x, n):
        data[y, x] = 1
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 21
data = np.zeros((n,n))
for i in range(n):
    data[i, 10] = 1
    data[10, i] = 1
for i in range(11):
    data[i + 10, i] = 1
    data[i, i + 10] = 1
    data[i, 10 - i] = 1
    data[i + 10 , 10 - i + 10] = 1
    
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 51
data = np.zeros((n,n))
for y in range(25):
    for x in range(y, n-y):
        data[y, x] = 1

i = 0
for y in range(25, n):
    for x in range(y - i, y+1):
        data[y, x] = 1
    i += 2
        
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
def dist(x, y, x_2, y_2):
    x_diff = abs(x_2 - x)
    y_diff = abs(y_2 - y)
    x_sqr = x_diff ** 2
    y_sqr = y_diff ** 2
    distance = (x_sqr + y_sqr) ** 0.5
    return distance

n = 501
r = n/2.0
data = np.zeros((n,n))
for x in range(n):
    for y in range(n):
        dist1 = dist(x, y, 0, 0)
        dist2 = dist(x, y, 501, 501)
        dist3 = dist(x, y, 0, 501)
        dist4 = dist(x, y, 501, 0)
        if (dist1 < r or dist2 < r or dist3 < r or dist4 < r):
            data[x, y] = 1
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 501
r = 160
data = np.zeros((n,n))
for x in range(n):
    for y in range(n):
        dist1 = dist(x, y, 150, 251)
        dist2 = dist(x, y, 350, 251)
        if (dist1 < r or dist2 < r):
            data[y, x] = 1
        if (dist1 < r and dist2 < r):
            data[y, x] = 0
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
H = np.loadtxt('topography_180x360_grid.txt')
```

```python
max_elevation = -10000

max_x = 0
max_y = 0

for x in range(360):
    for y in range(180):
        elevation = H[y, x]
        if elevation >= max_elevation:
            max_elevation = elevation
            max_x = x
            max_y = y

print(max_x, max_y)

for x in range(max_x - 10, max_x + 10):
    for y in range(max_y - 10, max_y + 10):
        H[x, y] = 10000
        
    
plt.imshow(H, interpolation='nearest')
plt.show()
```
